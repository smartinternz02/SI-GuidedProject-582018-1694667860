```python
import seaborn as sns
data=sns.load_dataset('iris')
```

```python
data.head()
```

```python
data.describe()
```

```python
sns.distplot(data['petal_length'],bins=30)
```

```python
sns.distplot(data['petal_length'],hist=False, bins=10)
```

```python
sns.relplot(x="petal_length",y="sepal_length",data=data,hue= "species")
```

```python
sns.relplot(x="petal_length",y="sepal_length",kind='line',data=data)
```

```python
sns.relplot(x="petal_length",y="sepal_length",kind='line',sort=False, data=data)
```

```python
sns.catplot(x="species", y="petal_length", data= data)
```

```python
sns.boxplot(x='species', y='sepal_length', data=data)
```

```python
sns.boxplot(x='species', y='sepal_width', data=data)
```
